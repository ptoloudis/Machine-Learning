The traffic arriving at a network gateway is bursty. The burst size x is geometrically distributed with the following pmf.

ùëì(ùë•)=(1‚àíùëù)ùë•‚àí1ùëù,ùë•=1,2,‚Ä¶,‚àû(53)
Compute the mean, variance, standard deviation, and coefficient of variation of the burst size. Plot the pmf and CDF for P = 0.2.


```python
import numpy as np
import matplotlib.pyplot as plt

p = 0.2

# Define the range of values for the burst size
x = np.arange(1, 21)

# Compute the pmf
pmf = (1-p)**(x-1) * p

# Compute the CDF
cdf = np.cumsum(pmf)

# Plot the pmf
plt.stem(x, pmf)
plt.title('PMF for Geometric Distribution (p=0.2)')
plt.xlabel('Burst Size')
plt.ylabel('Probability')
plt.show()

# Plot the CDF
plt.step(x, cdf)
plt.title('CDF for Geometric Distribution (p=0.2)')
plt.xlabel('Burst Size')
plt.ylabel('Probability')
plt.show()
```

Exercise 5: The response time of a computer system has an Erlang distribution with the following CDF:
$F(x)=1-{e^{-x/a}}{\sum_{i=0}^{m-1}\frac{(x/a)^i}{i!}}$
Find expressions for the pdf, mean, variance, mode, and coefficient of variation of the response time.

To find the pdf of the Erlang distribution, we take the derivative of the CDF:

$$f(x) = \frac{dF(x)}{dx} = \frac{e^{-x/a}}{a\sum_{i=0}^{m-1}\frac{(x/a)^i}{i!}}\left(\sum_{i=0}^{m-1}\frac{(x/a)^i}{i!} - \frac{(x/a)^m}{m!}\right)$$

The mean of the Erlang distribution is given by:

$$\mu = ma$$

The variance is given by:

$$\sigma^2 = m a^2$$

To find the mode of the distribution, we need to find the value of $x$ that maximizes the pdf. This occurs when the derivative of the pdf with respect to $x$ is equal to zero. Unfortunately, there is no closed-form solution for this equation, so we typically use numerical methods to find the mode.

The coefficient of variation is given by:

$$CV = \frac{\sigma}{\mu} = \frac{\sqrt{m}a}{ma} = \frac{1}{\sqrt{m}}$$

Note that the coefficient of variation decreases as $m$ increases, indicating that the distribution becomes less variable as the shape parameter increases.

```python
# Step 1: Import the necessary libraries and read the data from the file
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from scipy import stats

df = pd.read_csv("data_2.csv")

# Step 2: Create the feature matrix X and target variable y from the data
X = df.iloc[:, 0].values.reshape(-1, 1)  # Record Size
y = df.iloc[:, 1:].values              # Observations

# Step 3: Fit a linear regression model to the data using the scikit-learn library
reg = LinearRegression(fit_intercept=False).fit(X, y)

# Step 4: Evaluate the performance of the model using statistical measures like R-squared, p-value, and confidence intervals
# R-squared
r_squared = reg.score(X, y)

# p-value and confidence intervals
n = len(X)
k = 1     # We have only one feature, i.e., the slope coefficient
t_value = stats.t.ppf(0.975, n-k-1)
std_err = stats.tstd(y, axis=0) / np.sqrt(n)
conf_int = np.vstack((reg.coef_[0][0] - t_value*std_err, reg.coef_[0][0] + t_value*std_err)).T

print("R-squared:", r_squared)
print("Confidence Intervals:")
print(conf_int)

# Step 5: Use the model to make predictions and visualize the results
y_pred = reg.predict(X)

y1 = y.T[0]
plt.scatter(X, y1)
plt.plot(X, y_pred, color='red')
plt.xlabel('Record Size (k-bytes)')
plt.ylabel('Observation Time (seconds)')
plt.title('Observation Time vs Record Size')
plt.show()

y2 = y.T[1]
plt.scatter(X, y2)
plt.plot(X, y_pred, color='red')
plt.xlabel('Record Size (k-bytes)')
plt.ylabel('Observation Time (seconds)')
plt.title('Observation Time vs Record Size')
plt.show()

y3 = y.T[2]
plt.scatter(X, y3)
plt.plot(X, y_pred, color='red')
plt.xlabel('Record Size (k-bytes)')
plt.ylabel('Observation Time (seconds)')
plt.title('Observation Time vs Record Size')
plt.show()
```
