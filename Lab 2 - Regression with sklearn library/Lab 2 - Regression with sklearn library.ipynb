2. PolynomialFeatures: `PolynomialFeatures` is a preprocessing class in sklearn that generates a new feature matrix consisting of all polynomial combinations of the features up to a specified degree. For example, if you have a single feature X and you want to create polynomial features up to degree 3, you can use `PolynomialFeatures(degree=3)`. You can then use the transformed feature matrix with any regression model of your choice.

3. Ridge Regression: `Ridge` is a linear regression model that incorporates L2 regularization to prevent overfitting. You can use the `PolynomialFeatures` class to generate polynomial features and then use `Ridge` to perform polynomial regression with regularization.

4. Lasso Regression: `Lasso` is a linear regression model that incorporates L1 regularization. Like `Ridge`, you can use `PolynomialFeatures` to generate polynomial features and then use `Lasso` to perform polynomial regression with regularization.

5. ElasticNet Regression: `ElasticNet` is a linear regression model that incorporates both L1 and L2 regularization. You can use `PolynomialFeatures` to generate polynomial features and then use `ElasticNet` to perform polynomial regression with regularization.

To use these methods, you will typically need to first preprocess your data and create the polynomial features. You can then use any of the regression models mentioned above to fit your model and make predictions.

# Scikit-learn (sklearn) python library for machine learning

Scikit-learn (sklearn) is a popular Python library for machine learning. It provides a variety of regression algorithms that can be used for modeling continuous target variables. Some of the commonly used regression methods in sklearn are:

1. Linear Regression: Linear regression is a linear approach to modeling the relationship between a dependent variable and one or more independent variables. In sklearn, linear regression is implemented using the `LinearRegression` class.

2. Ridge Regression: Ridge regression is a regularized form of linear regression that is used to prevent overfitting. It introduces a regularization term to the objective function that penalizes large coefficients. In sklearn, ridge regression is implemented using the `Ridge` class.

3. Lasso Regression: Lasso regression is another regularized form of linear regression that is used to prevent overfitting. It introduces a regularization term to the objective function that penalizes the absolute value of the coefficients. In sklearn, lasso regression is implemented using the `Lasso` class.

4. ElasticNet Regression: ElasticNet regression is a combination of ridge and lasso regression that is used to prevent overfitting. It introduces a regularization term to the objective function that is a linear combination of the ridge and lasso penalties. In sklearn, ElasticNet regression is implemented using the `ElasticNet` class.

5. Support Vector Regression (SVR): Support vector regression is a non-linear regression method that is based on the idea of support vector machines (SVMs). In sklearn, SVR is implemented using the `SVR` class.

6. Decision Tree Regression: Decision tree regression is a non-linear regression method that is based on the idea of decision trees. In sklearn, decision tree regression is implemented using the `DecisionTreeRegressor` class.

7. Random Forest Regression: Random forest regression is an ensemble method that combines multiple decision trees to improve the accuracy of the model. In sklearn, random forest regression is implemented using the `RandomForestRegressor` class.

8. Gradient Boosting Regression: Gradient boosting regression is another ensemble method that combines multiple weak models to improve the accuracy of the model. In sklearn, gradient boosting regression is implemented using the `GradientBoostingRegressor` class.

9. K-Nearest Neighbors Regression (KNN): KNN regression is a non-parametric regression method that predicts the target value based on the k-nearest neighbors in the training set. In sklearn, KNN regression is implemented using the `KNeighborsRegressor` class.

# Probabilistic Regression Methods
Scikit-learn (sklearn) is a popular machine learning library in Python that provides a variety of tools for regression tasks, including probabilistic regression methods. Probabilistic regression methods are those that provide not only point estimates of the target variable, but also an estimate of the uncertainty associated with the prediction.

Here are some of the probabilistic regression methods provided by sklearn:

1. Bayesian Ridge Regression:
This method is based on Bayesian statistics and adds a regularization term to the least squares objective function to avoid overfitting. It provides a probabilistic estimate of the target variable using a Gaussian distribution.

2. Gaussian Process Regression:
Gaussian Process Regression is a non-parametric approach that provides a distribution over functions that could generate the target variable. It allows for flexible modeling of the data and provides a probabilistic estimate of the target variable.

3. Ridge Regression with Cross-Validation:
This is a variant of ridge regression that uses cross-validation to determine the optimal value of the regularization parameter. It provides a probabilistic estimate of the target variable using a Gaussian distribution.

4. Passive Aggressive Regression:
This is an online learning algorithm that can be used for regression tasks. It updates the model parameters incrementally as new data becomes available. It provides a probabilistic estimate of the target variable using a hinge loss function.

5. Theil-Sen Regression:
Theil-Sen Regression is a robust regression method that estimates the slope of the line that best fits the data. It provides a probabilistic estimate of the target variable using a Gaussian distribution.

These are just a few of the probabilistic regression methods provided by sklearn. Each method has its own strengths and weaknesses, and the choice of method will depend on the specific characteristics of the data and the goals of the analysis.

# sklearn kernel regression methods

Scikit-learn (sklearn) is a popular machine learning library in Python. It provides various regression models, including kernel regression models. In kernel regression, a kernel function is used to weigh the influence of neighboring data points on the predicted value. Here are some kernel regression methods available in sklearn:

1. KernelRidge: This method combines ridge regression with the kernel trick. It can handle both linear and nonlinear problems.

2. SVR (Support Vector Regression): This method uses support vector machines (SVM) to perform regression. It can handle both linear and nonlinear problems and is particularly useful when dealing with high-dimensional data.

3. GaussianProcessRegressor: This method uses Gaussian processes to perform regression. It can handle both linear and nonlinear problems and is particularly useful when dealing with small datasets.

4. KNeighborsRegressor: This method performs regression by finding the k-nearest neighbors of a data point and averaging their target values. It can handle both linear and nonlinear problems.

5. RadiusNeighborsRegressor: This method is similar to KNeighborsRegressor, but instead of using a fixed number of neighbors, it uses all neighbors within a fixed radius.

6. DecisionTreeRegressor: This method fits a decision tree to the data and uses it to make predictions. It can handle both linear and nonlinear problems.

7. MLPRegressor: This method uses a multilayer perceptron (MLP) neural network to perform regression. It can handle both linear and nonlinear problems and is particularly useful when dealing with large datasets.

These are some of the kernel regression methods available in sklearn. Each method has its own strengths and weaknesses, and the choice of method depends on the specific problem at hand.

## Generation of synthetic datsets for testing kernel regression methods
To create a synthetic noise dataset for testing kernel regression methods in Python, you can use the `make_regression` function from the scikit-learn library. 

Here's an example code snippet that generates a synthetic noise dataset with 1000 samples and 10 features, and adds Gaussian noise to the target variable:

```python
from sklearn.datasets import make_regression
import numpy as np
import  matplotlib.pyplot as plt

# Generate synthetic dataset with 1000 samples and 10 features
X, y = make_regression(n_samples=1000, n_features=10, noise=0.1)

# Add Gaussian noise to target variable
y += np.random.normal(0, 1, size=y.shape)
x = X[:,1]

plt.scatter(x,y)
```

In this example, `X` is a 1000 x 10 matrix of features, and `y` is a 1000 x 1 vector of target values with added Gaussian noise. You can adjust the parameters of `make_regression` to generate datasets with different sizes and levels of noise. 

Once you have generated the synthetic dataset, you can use it to test different kernel regression methods in Python. For example, you could use the `KernelRidge` class from scikit-learn to fit a kernel ridge regression model:

```python
from sklearn.kernel_ridge import KernelRidge

# Fit kernel ridge regression model
model = KernelRidge(kernel='rbf')
model.fit(X, y)

# Make predictions on new data
X_new = np.random.randn(10)
y_pred = model.predict(X_new.reshape(1, -1))
```

In this example, we are fitting a kernel ridge regression model with a radial basis function (RBF) kernel to the synthetic noise dataset. We then use the model to make predictions on a new data point `X_new`. ### Further data preparation - scaling.

SVM, KNN, and linear models are very sensitive to the scaling of the data. There are several methods to transform the data that yield more standard ranges. In this project, I will use the MinMaxScaler that shifts the data such that all features are exactly between 0 and 1.

In particular, I scaled only the non-dummy variables in training dataset, and combined the scaled sub data set back to the dummy variables in the training data set. The same steps have been followed to the test dataset. So the datasets are ready for initial modeling. 

```python
Xm_train.columns
```

```python
#data transform
def scale (X):
    ### non scaled dataframe X --> scaled dataframe X
    #seperate dummy varialbes and nondummy variables for X_train and X_test
    X_nondummy = X[['Fedu', 'Medu', 'absences',  'age',
       'failures', 'famrel', 'freetime', 'goout',
       'health', 'studytime', 'traveltime', 'Aalc']]
    X_dummy = X[['Pstatus','activities', 'address','famsize', 'famsup','higher', 'internet', 'nursery', 'romantic', 'school',
       'schoolsup', 'sex', 'Fjob_health',
       'Fjob_other', 'Fjob_services', 'Fjob_teacher', 'Mjob_health',
       'Mjob_other', 'Mjob_services', 'Mjob_teacher', 'reason_home',
       'reason_other', 'reason_reputation', 'guardian_mother',
       'guardian_other']]
    #select scaler
    scaler = MinMaxScaler()
    #scale nondummy variables
    scaler.fit(X_nondummy)
    X_nondummy_scaled = scaler.transform(X_nondummy)
    #concatenate scaled nondummy with nonscaled dummy variables
    X_scaled = np.concatenate((X_nondummy_scaled, X_dummy.values), axis=1)
    return X_scaled
#Apply to Xm_train, Xm_test, Xp_train, Xp_test
Xm_train_scaled = scale(Xm_train)
Xm_test_scaled = scale(Xm_test)
Xp_train_scaled = scale(Xp_train)
Xp_test_scaled = scale(Xp_test)
```
