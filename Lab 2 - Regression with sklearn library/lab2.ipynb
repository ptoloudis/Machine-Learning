# Exercise 1: Approximate the function  ùëì(ùë•)=1/+ùë•^2 in the interval  [-5,5]
using the sklearn routines PolynomialFeatures without regularization and with regularization Lasso, Ridge, and ElasticNet.

Import the Library

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

Define the function to approximate

```python
def f(x):
    return 1/(1 + x**2)
```

Generate random data in the interval[-5,5]

```python
np.random.seed(0)
x_train = np.random.uniform(-5, 5, size=(100,))
y_train = f(x_train) + np.random.normal(scale=0.1, size=x_train.shape)
degrees = [3, 7, 15, 21]
```

Fit the data using PolynomialFeatures without regularization

```python
plt.figure(figsize=(16, 8))
for i, degree in enumerate(degrees):
    ax = plt.subplot(2, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    # Create a polynomial feature transformer and fit the data
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    linear_regression = LinearRegression()
    pipeline = make_pipeline(polynomial_features, linear_regression)
    pipeline.fit(x_train[:, np.newaxis], y_train)

    # Evaluate the model and plot the results
    x_plot = np.linspace(-5, 5, 100)
    y_plot = pipeline.predict(x_plot[:, np.newaxis])
    plt.plot(x_train, y_train, 'b.', markersize=10)
    plt.plot(x_plot, y_plot, 'r-', label='Degree %d' % degree)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim((-0.1, 1.1))
    plt.legend(loc='best')
```

Fit the data using Lasso regularization


```python
for i, degree in enumerate(degrees):
    ax = plt.subplot(2, len(degrees), i + 1 + len(degrees))
    plt.setp(ax, xticks=(), yticks=())

    # Create a polynomial feature transformer and fit the data
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    linear_regression = Lasso(alpha=0.01)
    pipeline = make_pipeline(polynomial_features, linear_regression)
    pipeline.fit(x_train[:, np.newaxis], y_train)

    # Evaluate the model and plot the results
    x_plot = np.linspace(-5, 5, 100)
    y_plot = pipeline.predict(x_plot[:, np.newaxis])
    plt.plot(x_train, y_train, 'b.', markersize=10)
    plt.plot(x_plot, y_plot, 'r-', label='Degree %d' % degree)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim((-0.1, 1.1))
    plt.legend(loc='best')
```

Fit the data using Ridge regularization

```python
for i, degree in enumerate(degrees):
    ax = plt.subplot(2, len(degrees), i + 1 + len(degrees))
    plt.setp(ax, xticks=(), yticks=())

    # Create a polynomial feature transformer and fit the data
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    linear_regression = Ridge(alpha=0.01)
    pipeline = make_pipeline(polynomial_features, linear_regression)
    pipeline.fit(x_train[:, np.newaxis], y_train)

    # Evaluate the model and plot the results
    x_plot = np.linspace(-5, 5, 100)
    y_plot = pipeline.predict(x_plot[:, np.newaxis])
    plt.plot(x_train, y_train, 'b.', markersize=10)
    plt.plot(x_plot, y_plot, 'r-', label='Degree %d' % degree)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim((-0.1, 1.1))
    plt.legend(loc='best')
```

Fit the data using ElasticNet regularization

```python
for i, degree in enumerate(degrees):
    ax = plt.subplot(2, len(degrees), i + 1 + len(degrees))
    plt.setp(ax, xticks=(), yticks=())

    # Create a polynomial feature transformer and fit the data
    polynomial_features = PolynomialFeatures(degree=degree, include_bias=False)
    linear_regression = ElasticNet(alpha=0.01, l1_ratio=0.5)
    pipeline = make_pipeline(polynomial_features, linear_regression)
    pipeline.fit(x_train[:, np.newaxis], y_train)

    # Evaluate the model and plot the results
    x_plot = np.linspace(-5, 5, 100)
    y_plot = pipeline.predict(x_plot[:, np.newaxis])
    plt.plot(x_train, y_train, 'b.', markersize=10)
    plt.plot(x_plot, y_plot, 'r-', label='Degree %d' % degree)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim((-0.1, 1.1))
    plt.legend(loc='best')
```

# Exercise 2: Identify a regression noise dataset from UCI repository https://archive.ics.uci.edu/ml/datasets.php and apply on it Bayesian Ridge Regression, Gaussian Process Regression, and Ridge Regression with Cross-Validation.

I choice the  "Energy efficiency" dataset

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import BayesianRidge, RidgeCV
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
from sklearn.model_selection import train_test_split
```

Load the dataset

```python
data = pd.read_excel('data/ENB2012_data.xlsx')
```

Split the data into input features (X) and output variable (y)

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

Split the data into training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Fit to the Bayesian Ridge Regrassion model  

```python
bay = BayesianRidge(compute_score=True)
bay.fit(X_train, y_train)
bay.score(X_test, y_test)
```

Fit to the Gaussian Process Regression model

```python
kernel = RBF()
gau = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10, normalize_y=True)
gau.fit(X_train, y_train)
gau.score(X_test, y_test)
```

Fit Ridge Regression with Cross-Validasion model

```python
cvm = RidgeCV(alphas=[0.1, 1.0, 10.0], cv=5)
cvm.fit(X_train, y_train)
cvm.score(X_test, y_test)
```

Print the score for each model

```python
print('BayesianRidge: ', bay.score(X_test, y_test))
print('GaussianProcessRegressor: ', gau.score(X_test, y_test))
print('RidgeCV: ', cvm.score(X_test, y_test))
```

# Exercise 5: This project will identify the factors among a student‚Äôs socioeconomic status, study attitude, school that impact the students‚Äô school performance most, especially to check if alcohol use will affect students‚Äô grades, with support of real-world data. In addition, it compares the machine learning techniques KNNs, SVR, Linear Regression and Trees for regression problems.

Import the librarys

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

Load the data

```python
data = pd.read_csv("data/student-mat.csv", delimiter=";")
# data = data[(data["school"] == "GP")]
X = data.iloc[:, :-3]
y = data.iloc[:, -3]
```

Convert categorical variables to numerical using LabelEncoder

```python
categorical_cols = ["school", "sex", "address", "famsize", "Pstatus", "Mjob", "Fjob", "reason", "guardian", "schoolsup", "famsup", "paid", "activities", "nursery", "higher", "internet", "romantic"]
le = LabelEncoder()
for col in categorical_cols:
    X[col] = le.fit_transform(X[col])
```

One-hot encode nominal categorical variables

```python
X = pd.get_dummies(X, columns=["Medu", "Fedu", "traveltime", "studytime", "failures", "famrel", "freetime", "goout", "Dalc", "Walc", "health"])
```

Split data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

Standardize features

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

KNN regression

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_pred)
```

SVR

```python
svr = SVR(kernel='linear')
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_test)
svr_mse = mean_squared_error(y_test, svr_pred)
```

Linear Regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_reg_pred = lin_reg.predict(X_test)
lin_reg_mse = mean_squared_error(y_test, lin_reg_pred)
```

Ridge Regression with Cross-Validation

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)
ridge_mse = mean_squared_error(y_test, ridge_pred)
```

Decision Tree Regression

```python
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train, y_train)
tree_reg_pred = tree_reg.predict(X_test)
tree_reg_mse = mean_squared_error(y_test, tree_reg_pred)
```

Print all

```python
print('KNN: ', knn_mse)
print('SVR: ', svr_mse)
print('Linear Regression: ', lin_reg_mse)
print('Ridge: ', ridge_mse)
print('Decision Tree: ', tree_reg_mse)
```

# Exercise7: Choose a dataset from UCI repository and apply the above analysis.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.datasets import make_regression
from sklearn.linear_model import Ridge, Lasso
from sklearn.svm import SVR
```

Load the dataset

```python
X, y = make_regression(n_samples=1000, n_features=10, noise=0.1)
y += np.random.normal(0, 1, size=y.shape)

scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y.reshape(-1, 1))
```

Split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Train and Eavaluate

Grandient Boosting Machines

```python
gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=1, random_state=0, loss='ls')
gbr.fit(X_train, y_train)
gbr_pred = gbr.predict(X_test)
gbr_mse = mean_squared_error(y_test, gbr_pred)
```

Ridge Regression

```python
ridge = Ridge(alpha=0.4)
ridge.fit(X_train, y_train)
ridge_pred = ridge.predict(X_test)
ridge_mse = mean_squared_error(y_test, ridge_pred)
```

SVR

```python
svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_test)
svr_mse = mean_squared_error(y_test, svr_pred)
```

Lasso Regression

```python
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
lasso_pred = lasso.predict(X_test)
lasso_mse = mean_squared_error(y_test, lasso_pred)
```

Random Forest

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=2, random_state=0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_mse = mean_squared_error(y_test, rf_pred)
```

Decision Tree

```python
dt = DecisionTreeRegressor(max_depth=10)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)
dt_mse = mean_squared_error(y_test, dt_pred)
```

KNN

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
knn_mse = mean_squared_error(y_test, knn_pred)
```

Print the result

```python
print('GradientBoostingRegressor: ', gbr_mse, gbr.score(X_test, y_test))
print('Ridge: ', ridge_mse, ridge.score(X_test, y_test))
print('SVR: ', svr_mse, svr.score(X_test, y_test))
print('Lasso: ', lasso_mse, lasso.score(X_test, y_test))
print('RandomForestRegressor: ', rf_mse, rf.score(X_test, y_test))
print('DecisionTreeRegressor: ', dt_mse, dt.score(X_test, y_test))
print('KNeighborsRegressor: ', knn_mse, knn.score(X_test, y_test))
```
