2. Load the dataset into a Pandas DataFrame and preprocess it as needed:

```python
df = pd.read_csv('data.zip')
df = df.dropna() # remove any rows with missing data
df['utc_timestamp'] = pd.to_datetime(df['utc_timestamp']) # convert the timestamp to a datetime object
df = df.set_index('utc_timestamp') # set the timestamp as the index
```

3. Split the data into training and testing sets:

```python
split_index = int(len(df)*0.8) # use 80% of the data for training
train_data = df.iloc[:split_index]
test_data = df.iloc[split_index:]
```

4. Normalize the data to a range between 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)
```

5. Create sequences of data to use as input and output to the LSTM model:

```python
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i])
        y.append(data[i])
    return np.array(X), np.array(y)

seq_length = 24 # use 24 hours of data to predict the next hour
train_X, train_y = create_sequences(train_data, seq_length)
test_X, test_y = create_sequences(test_data, seq_length)
```

6. Define the LSTM model:

```python
model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')
```

7. Train the model on the training data:

```python
history = model.fit(train_X, train_y, epochs=50, batch_size=32, validation_data=(test_X, test_y), verbose=2, shuffle=False)
```

8. Evaluate the model on the testing data:

```python
test_loss = model.evaluate(test_X, test_y, verbose=0)
print('Test loss:', test_loss)
```
