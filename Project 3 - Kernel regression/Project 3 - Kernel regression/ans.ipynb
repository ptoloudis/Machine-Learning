```python
import numpy as np
from math import exp,sqrt,pi
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import sklearn
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
def fit(test_X, train_X, train_y, bandwidth=1.0, kn='box'):
    kernels = {
        'box': lambda x: 1/2 if (x<=1 and x>=-1) else 0,
        'gs': lambda x: 1/sqrt(2*pi)*exp(-x**2/2),
        'ep': lambda x: 3/4*(1-x**2) if (x<=1 and x>=-1) else 0
    }
    predict_y = []
    for entry in test_X:
        nks = [np.sum((j-entry)**2)/bandwidth for j in train_X]
        ks = [kernels['box'](i) for i in nks]
        dividend = sum([ks[i]*train_y[i] for i in range(len(ks))])
        divisor = sum(ks)
        predict = dividend/divisor
        predict_y.extend([predict])
    return np.array(predict_y)[:,np.newaxis]
```

```python
data = pd.read_csv('df.csv')
data.head()
x_data= data.x
y_data= data.y
m_data = data.m
plt.plot(x_data,y_data)
```

```python
result = fit(x_data,x_data,y_data,bandwidth=0.1,kn='box')
plt.plot(x_data,result)
```

The following code generates a 3d dataset in matlab. Write a python script that performs Gaussian Kernel Regression on this dataset.

```python
# Define the range as a row vector.
x = np.arange(-10, 10.2, 0.2)

# Create a 2D grid holding the x1 values at every point for the grid.
# Every *row* of this matrix will be identical.
X1 = np.ones((len(x), len(x)))
for i in range(len(x)):
    X1[i, :] = x

# Create a 2D grid holding the x2 values at every point for the grid.
# Every *column* of this matrix will be identical.
X2 = np.ones((len(x), len(x)))
for i in range(len(x)):
    X2[:, i] = x

# Unwind the two grids into a single list of 2D coordinates.
X = np.column_stack((X1.ravel(), X2.ravel()))

# Evaluate the sombrero function over every input coordinate.
A = np.sqrt(X[:, 0]**2 + X[:, 1]**2)
Z_orig = np.sin(A) / A

# The sombrero function evaluates to NaN at (0, 0), but the correct
# value is 1.
Z_orig[np.isnan(Z_orig)] = 1

# Plot the sombrero function using the 'plot_surface' function.
# In order to plot the Z values, we need to reshape the vector
# back into a 2D matrix.
fig = plt.figure()
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X1, X2, Z_orig.reshape(len(x), len(x)))
ax.set_title('Original Function')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-0.5, 1.0])

# Add random noise to the data.
Z = Z_orig + (0.1 * np.random.randn(*Z_orig.shape))

# Plot the noisy sombrero using the 'plot_surface' function.
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X1, X2, Z.reshape(len(x), len(x)))
ax.set_title('Noisy Data')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-0.5, 1.0])

plt.show()
```

```python
# Define the Gaussian kernel function.
def gaussian_kernel(x, y, sigma=1.0):
    return np.exp(-np.sum((x - y)**2) / (2 * sigma**2))

# Define the Gaussian kernel regression function.
def gaussian_kernel_regression(x_train, y_train, x_test, sigma=1.0):
    y_pred = np.zeros(len(x_test))
    for i, x in enumerate(x_test):
        k = np.array([gaussian_kernel(x, x_train[j], sigma) for j in range(len(x_train))])
        y_pred[i] = np.sum(k * y_train) / np.sum(k)
    return y_pred

# Generate the dataset.
x = np.linspace(-5, 5, 100)
y = np.sin(x) + np.random.normal(0, 0.1, size=x.shape)

# Split the dataset into training and testing sets.
train_indices = np.random.choice(len(x), size=20, replace=False)
x_train, y_train = x[train_indices], y[train_indices]
x_test, y_test = np.delete(x, train_indices), np.delete(y, train_indices)

# Perform Gaussian kernel regression.
y_pred = gaussian_kernel_regression(x_train, y_train, x_test, sigma=0.5)

# Plot the results.
plt.scatter(x_train, y_train, color='red', label='Training Data')
plt.scatter(x_test, y_test, color='blue', label='Testing Data')
plt.plot(x_test, y_pred, color='green', label='Predicted Values')
plt.legend()
plt.show()
```

```python
# Generate the dataset.
x = np.arange(-10, 10.2, 0.2)
X1 = np.ones((len(x), len(x)))
for i in range(len(x)):
    X1[i, :] = x
X2 = np.ones((len(x), len(x)))
for i in range(len(x)):
    X2[:, i] = x
X = np.column_stack((X1.ravel(), X2.ravel()))
A = np.sqrt(X[:, 0]**2 + X[:, 1]**2)
Z_orig = np.sin(A) / A
Z_orig[np.isnan(Z_orig)] = 1

# Add random noise to the data.
Z = Z_orig + (0.1 * np.random.randn(*Z_orig.shape))

# Generate a new set of test points on the x1-x2 plane
x_test = np.arange(-10, 10.1, 0.1)
X1_test = np.ones((len(x_test), len(x_test)))
for i in range(len(x_test)):
    X1_test[i, :] = x_test
X2_test = np.ones((len(x_test), len(x_test)))
for i in range(len(x_test)):
    X2_test[:, i] = x_test
X_test = np.column_stack((X1_test.ravel(), X2_test.ravel()))
```

```python
# Apply the k-NN regression function to the dataset.
k = 10
knn = KNeighborsRegressor(n_neighbors=k)
knn.fit(X, Z)
```

```python
# Predict the function values at the test points using k-NN regression
Z_pred = knn.predict(X_test)
```

```python
# Reshape the values into a grid
Z_orig_2d = Z_orig.reshape(X1.shape)
Z_pred_grid = Z_pred.reshape(len(x_test), len(x_test))

# Plot the original sombrero surface and the predicted surface using k-NN regression
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z_orig_2d, cmap='jet')
ax1.set_title('Original sombrero surface')
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X1_test, X2_test, Z_pred_grid, cmap='jet')
ax2.set_title('Predicted surface using k-NN regression')
plt.show()
```

```python
bandwidth = 0.1

# Fit the local polynomial regression to the sombrero function
Z_loess = lowess(Z_orig, X[:,0], frac=bandwidth, return_sorted=False)

# Reshape the output to the shape of X1 and X2
Z_loess_2d = Z_loess.reshape(X1.shape)
```

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X1, X2, Z_loess_2d, cmap='jet')
ax1.set_title('Local polynomial regression surface')
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X1, X2, Z_orig_2d, cmap='jet')
ax2.set_title('Original sombrero surface')
plt.show()
```
